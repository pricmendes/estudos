```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando os dados
ratings = pd.read_csv(r"ratings.csv")
movies = pd.read_csv(r"movies.csv")
tags = pd.read_csv(r"tags.csv")
links = pd.read_csv(r"links.csv")
```

```python
# Junção dos datasets (exemplo com ratings e movies)
df = pd.merge(ratings, movies, on='movieId')

```

```python
# Visualizando as primeiras linhas do DataFrame
print(df.head())
```

```python
# a. Gráfico de barras: número de avaliações dos 10 filmes mais populares
popular_movies = df.groupby('title')['userId'].count().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=popular_movies.index, y=popular_movies.values)
plt.xticks(rotation=45)
plt.title('10 Filmes Mais Avaliados')
plt.show()

```

```python
# b. Histograma: distribuição das notas
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='rating', bins=5, kde=True)
plt.title('Distribuição das Notas')
plt.show()

```

```python
# ... (continuar com os outros gráficos)

# d. Gráfico de linha: evolução da média de avaliações ao longo do tempo
# (assumindo que a coluna 'timestamp' em ratings represente o tempo)
df['year'] = pd.to_datetime(df['timestamp'], unit='s').dt.year
yearly_avg = df.groupby('year')['rating'].mean()
plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_avg.index, y=yearly_avg.values)
plt.title('Evolução da Média de Avaliações ao Longo do Tempo')
plt.show()

```
