```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arquivo = r"C:\Users\prisc\OneDrive\Documentos\aPRI\ESTUDOS\DADOS\UNICESUMAR\VISUALIZAÇÃO DE DADOS\iris_data_flowerszip\IRIS.csv"
df = pd.read_csv(arquivo)

contagem_linhas = len(df.index)
contagem_colunas = len(df.columns)

print(f"Qtd. Colunas:", [contagem_colunas])
print(f"Qtd. Linhas:", [contagem_linhas])
# df.shape # aqui demonstra em tuplas a qtd de linhas e colunas
df.head()
```

```python
# Listagem  das espécies no dataset (valor único)
lista_especies = df['species'].unique()
print("Espécies observadas no dataset:")
for especies in lista_especies:
    print(especies)
```

```python
# Média de largura e comprimento da sépala para cada espécie
sepala_media = df.groupby('species')[['sepal_length', 'sepal_width']].mean()
sepala_media

```

```python
# Configurando o estilo visual
sns.set(style="whitegrid")

# Traçando o gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='viridis', s=100)
plt.title('Dispersão do Comprimento e Largura da Sépala por Espécie')
plt.xlabel('Comprimento da Sépala (cm)')
plt.ylabel('Largura da Sépala (cm)')
plt.legend(title='Espécie')
plt.show()
```

```python
# nota-se pelo gráfico que a Iris-setosa está mais distinta das outras espécies tornando mais fácil de identificá-la
# Já as espécies Iris-versicolor e Iris-virginica estão mais próximas e tornando assim mais difícil distingui-las.
```

```python
# Configurando o estilo com o fundo branco gradeado
sns.set(style="whitegrid")

# Criando uma figura com 4 subplots (2x2) duas linhas duas colunas cada um associado a variavel de comprimento/largura 
plt.figure(figsize=(12, 10))

# Histograma para o comprimento da sépala
plt.subplot(2, 2, 1) # será exibido na primeira matriz dos subplots
sns.histplot(df['sepal_length'], kde=True, color='blue') #coluna e cor azul
plt.title('Comprimento da Sépala')

# Histograma para a largura da sépala
plt.subplot(2, 2, 2) # segundo subplot cor verde
sns.histplot(df['sepal_width'], kde=True, color='green')
plt.title('Largura da Sépala')

# Histograma para o comprimento da pétala
plt.subplot(2, 2, 3) # terceiro subplot cor vermelho
sns.histplot(df['petal_length'], kde=True, color='red')
plt.title('Comprimento da Pétala')

# Histograma para a largura da pétala
plt.subplot(2, 2, 4) # quarto subplo cor rosa
sns.histplot(df['petal_width'], kde=True, color='purple')
plt.title('Largura da Pétala')

# Ajustar o layout para evitar sobreposição espaçamento dos blocos
plt.tight_layout()
plt.show()
```
